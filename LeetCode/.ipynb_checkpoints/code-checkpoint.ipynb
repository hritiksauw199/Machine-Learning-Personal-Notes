# 1. Two Sum

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.



```python
# brute force
def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i,j]
                
#optimized
def twoSum(self, nums: List[int], target: int) -> List[int]:
    num_to_index = {}
        
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_to_index:
            return [num_to_index[complement], i]
        num_to_index[num] = i
```

# 2. Add Two numbers

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.



```python
def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy_head = ListNode(0) # node starts at 0 for initial one
        current = dummy_head # pointer at current node
        carry = 0 # initial carry is 0
        
        # Iterate while either list still has nodes or there is a carry
        while l1 or l2 or carry:
            # Get the current values or default to 0
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0
            
            # Compute the sum
            total = val1 + val2 + carry
            
            # Update the carry for the next iteration
            carry = total // 10 # returns the quotient 
            
            # Append the last digit of total to the result list
            current.next = ListNode(total % 10) # returns the remainder
            current = current.next
            
            # Move to the next nodes in the lists
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
            
        return dummy_head.next
```
